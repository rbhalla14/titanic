```python
import pandas as pd


df = pd.read_csv("train.csv")
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#성별로 분류한 생존자/사망자

```python
sns.barplot(x = 'Sex', y = 'Survived', data = df)
```

남자는 약 20% 여자는 약 75%가 생존했다.


## 좌석 등급으로 분류한 생존 확률

```python
sns.barplot(x = 'Pclass', y = 'Survived', data = df)
```

1등석 사람이 제일 많이 살아남았고
3등석 사람의 생존율이 제일 낮았다
돈이 많은 사람들이 더 많이 살아남은 것 같다.

## 성별+좌석 등급으로 분류한 생존 확률

```python
sns.barplot(x = 'Pclass', y = 'Survived', hue='Sex', data = df)
```

성별과 객실 등급을 합친 표다.
여기서도 여자가 남자보다 생존율이 높은점을 확인할 수 있다.
또한 모든 객실 등급에서도 여자가 남자보다 생존율이 높았다.
따라서 1등석 여자가 생존율이 가장 높았고, 3등석 남자가 생존율이 가장 낮았다.
또한 3등석 여자가 1등석 남자보다 생존율이 높다는 점도 눈여겨볼만하다.
